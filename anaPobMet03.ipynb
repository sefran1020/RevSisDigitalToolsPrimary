```python
# %% [markdown]
# # Systematic Review Analysis: Interactive Digital Tools in Mathematics (Primary Education)
# ## Population, Methodology, and Results Analysis

# %% [markdown]
# ### 1. Initial Setup and Configuration

# %%
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from collections import Counter
import re
from scipy.stats import chi2_contingency
import networkx as nx
from pathlib import Path
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import plotly.graph_objects as go
# import plotly.express as px # Not used directly, can be removed if desired
from plotly.subplots import make_subplots
import logging
from collections import defaultdict, Counter # Agrega defaultdict aqu
# Basic Logging Configuration
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Configure plot style and appearance
plt.style.use('seaborn-v0_8-darkgrid') # Use a seaborn style
# sns.set_palette("husl") # Set a specific color palette (optional)
plt.rcParams['figure.figsize'] = (12, 8) # Default figure size
plt.rcParams['font.size'] = 12         # Default font size

# --- Define Output Directory for Figures ---
FIGURES_DIR = Path("pobMetFiguras")
FIGURES_DIR.mkdir(exist_ok=True) # Create the directory if it doesn't exist
logging.info(f"Figures will be saved to: {FIGURES_DIR.resolve()}")

# --- NLTK Resource Download ---
# Download necessary NLTK data (punkt for tokenization, stopwords)
try:
    _ = nltk.data.find('tokenizers/punkt')
    logging.debug("NLTK 'punkt' resource found.")
except nltk.downloader.DownloadError:
    logging.info("Downloading NLTK 'punkt' resource...")
    nltk.download('punkt', quiet=True)
    logging.info("'punkt' downloaded.")
except Exception as e:
     logging.warning(f"Could not verify/download NLTK 'punkt': {e}")

try:
    _ = nltk.data.find('corpora/stopwords')
    logging.debug("NLTK 'stopwords' resource found.")
except nltk.downloader.DownloadError:
    logging.info("Downloading NLTK 'stopwords' resource...")
    nltk.download('stopwords', quiet=True)
    logging.info("'stopwords' downloaded.")
except Exception as e:
     logging.warning(f"Could not verify/download NLTK 'stopwords': {e}")

# Define stopwords set (using English, adjust if needed)
stop_words = set(stopwords.words('english'))

# %% [markdown]
# ### 2. Data Loading

# %%
# Load the dataset
CSV_FILE = 'poblacionMetodologia.csv'
try:
    df = pd.read_csv(CSV_FILE, sep=';', encoding='utf-8')
    logging.info(f"Data loaded from '{CSV_FILE}': {df.shape[0]} records, {df.shape[1]} columns")
    # Basic cleaning: strip whitespace from column names
    df.columns = df.columns.str.strip()
    print("\nColumns available:")
    for i, col in enumerate(df.columns, 1):
        print(f"{i}. {col}")
except FileNotFoundError:
    logging.error(f"Error: File not found at '{CSV_FILE}'. Please check the path.")
    # Exit or raise error if file is crucial
    raise SystemExit(f"Error: File not found - {CSV_FILE}")
except Exception as e:
    logging.error(f"An error occurred while loading the CSV: {e}")
    raise SystemExit("Error loading data.")

# %% [markdown]
# ### 3. Exploratory Data Analysis (Basic Info)

# %%
# Display basic info and check data types
print("\nDataFrame Info:")
df.info()

# Convert 'Year' to numeric, coercing errors, and describe
if 'Year' in df.columns:
    df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
    print("\nPublication Year Statistics:")
    print(df['Year'].describe().apply(lambda x: f"{x:.2f}")) # Format output
else:
    logging.warning("Column 'Year' not found.")

# %% [markdown]
# ### 4. Geographic Analysis

# %%
# --- Geographic Distribution ---
def extract_first_country(country_str):
    """Extracts the first country from a potentially multi-country string."""
    if pd.isna(country_str):
        return None
    countries = [c.strip() for c in str(country_str).split(',')]
    return countries[0] if countries else None

if 'Country' in df.columns:
    df['FirstCountry'] = df['Country'].apply(extract_first_country)
    country_counts = df['FirstCountry'].dropna().value_counts()

    if not country_counts.empty:
        logging.info(f"Found {len(country_counts)} unique countries.")
        # Create bar chart for geographic distribution
        plt.figure(figsize=(15, 8)) # Adjusted size
        bars = plt.bar(country_counts.index, country_counts.values, color='skyblue', edgecolor='navy')
        plt.xticks(rotation=60, ha='right', fontsize=10) # Adjusted rotation and size
        plt.xlabel('Country (First Mentioned)')
        plt.ylabel('Number of Studies')
        plt.title('Geographic Distribution of Studies (by First Country)', fontsize=16, fontweight='bold')

        # Add value labels on bars
        for bar in bars:
            height = bar.get_height()
            if height > 0: # Only label bars with height > 0
                 plt.text(bar.get_x() + bar.get_width()/2., height,
                          f'{int(height)}', ha='center', va='bottom', fontsize=9)

        plt.tight_layout()
        plt.savefig(FIGURES_DIR / 'distribucion_geografica.png', dpi=300, bbox_inches='tight')
        logging.info(f"Saved geographic distribution plot to {FIGURES_DIR / 'distribucion_geografica.png'}")
        plt.show()
    else:
        logging.warning("No valid country data found for geographic distribution plot.")
else:
    logging.warning("Column 'Country' not found. Skipping geographic analysis.")

# --- Regional Analysis ---
def get_region(country):
    """Assigns a region based on predefined country lists."""
    if pd.isna(country): return 'Not Specified' # Handle potential None from extract_first_country
    # Define country lists within the function or load from external source
    asian_countries = ['China', 'Taiwan', 'Japan', 'Indonesia', 'India', 'Malaysia', 'United Arab Emirates', 'Korea'] # Added Korea
    european_countries = ['Slovakia', 'Italy', 'Romania', 'Spain', 'Hungary', 'Poland', 'Serbia', 'The Netherlands', 'Ukraine', 'Greece', 'Portugal', 'Germany', 'UK'] # Added more
    american_countries = ['Peru', 'Mexico', 'Colombia', 'USA', 'Canada', 'Brazil', 'Chile'] # Added more

    country_lower = country.lower() # Compare lowercase for consistency

    if any(c.lower() in country_lower for c in asian_countries): return 'Asia'
    if any(c.lower() in country_lower for c in european_countries): return 'Europe'
    if any(c.lower() in country_lower for c in american_countries): return 'America'
    # Add other regions like Africa, Oceania if applicable
    return 'Other/Not Specified'

if 'FirstCountry' in df.columns:
    df['Region'] = df['FirstCountry'].apply(get_region)
    region_counts = df['Region'].value_counts()

    if not region_counts.empty:
        logging.info("Generating regional distribution plot...")
        # Pie chart for regions
        plt.figure(figsize=(10, 8))
        colors = plt.cm.Pastel1(np.linspace(0, 1, len(region_counts))) # Use a colormap
        patches, texts, autotexts = plt.pie(region_counts.values,
                                            autopct='%1.1f%%',
                                            colors=colors,
                                            startangle=90,
                                            pctdistance=0.85) # Adjust label distance
        # Improve text visibility
        for text in texts: text.set_fontsize(10); text.set_color('black')
        for autotext in autotexts: autotext.set_color('black'); autotext.set_fontsize(9)

        plt.title('Regional Distribution of Studies', fontsize=16, fontweight='bold')
        # Add legend instead of labels on slices for clarity if many regions
        plt.legend(region_counts.index, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
        plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
        plt.tight_layout()
        plt.savefig(FIGURES_DIR / 'distribucion_regional.png', dpi=300, bbox_inches='tight')
        logging.info(f"Saved regional distribution plot to {FIGURES_DIR / 'distribucion_regional.png'}")
        plt.show()
    else:
        logging.warning("No region data found for regional distribution plot.")
else:
    logging.warning("Column 'FirstCountry' not generated. Skipping regional analysis.")

# %% [markdown]
# ### 5. Temporal Analysis

# %%
# --- Temporal Trend ---
if 'Year' in df.columns and not df['Year'].isnull().all():
    year_counts = df['Year'].dropna().astype(int).value_counts().sort_index()

    if not year_counts.empty:
        logging.info("Generating temporal trend plot...")
        # Ensure index is numeric for plotting
        year_counts.index = pd.to_numeric(year_counts.index)

        plt.figure(figsize=(12, 7)) # Adjusted size
        plt.plot(year_counts.index, year_counts.values, marker='o', linewidth=2.5, markersize=8, color='teal') # Changed color
        plt.fill_between(year_counts.index, year_counts.values, alpha=0.2, color='teal')
        plt.xlabel('Year')
        plt.ylabel('Number of Studies')
        plt.title('Temporal Trend of Publications', fontsize=16, fontweight='bold') # Adjusted title
        plt.grid(True, axis='y', linestyle='--', alpha=0.6) # Grid on y-axis only
        plt.xticks(ticks=np.arange(year_counts.index.min(), year_counts.index.max() + 1), rotation=45) # Ensure all years shown if reasonable

        # Add value labels
        for year, count in year_counts.items():
             plt.text(year, count + 0.15, str(count), ha='center', va='bottom', fontsize=10, fontweight='bold')

        plt.tight_layout()
        plt.savefig(FIGURES_DIR / 'tendencia_temporal.png', dpi=300, bbox_inches='tight')
        logging.info(f"Saved temporal trend plot to {FIGURES_DIR / 'tendencia_temporal.png'}")
        plt.show()
    else:
        logging.warning("No valid year data found for temporal trend plot.")
else:
    logging.warning("Column 'Year' not found or is empty. Skipping temporal analysis.")


# %% [markdown]
# ### 6. Methodology Analysis

# %%
# --- Methodology Analysis ---
# Check if columns exist before using them
method_cols = {
    'Quantitative': 'Quantitative Methods',
    'Qualitative': 'Qualitative Methods',
    'Mixed': 'Mixed Methods'
}
methods_analysis = {}
valid_method_cols_found = False

for method_name, col_name in method_cols.items():
    if col_name in df.columns:
        methods_analysis[method_name] = df[col_name].notna().sum()
        valid_method_cols_found = True
        logging.info(f"Column '{col_name}' found. Count for {method_name}: {methods_analysis[method_name]}")
    else:
        methods_analysis[method_name] = 0 # Assign 0 if column is missing
        logging.warning(f"Methodology column '{col_name}' not found.")

if valid_method_cols_found and any(v > 0 for v in methods_analysis.values()):
    logging.info("Generating methodology distribution plot...")
    # Create bar chart for methodologies
    methods_df = pd.DataFrame({
        'Methodology': list(methods_analysis.keys()),
        'Studies': list(methods_analysis.values())
    }).sort_values('Studies', ascending=False)

    plt.figure(figsize=(10, 7)) # Adjusted size
    bars = plt.bar(methods_df['Methodology'], methods_df['Studies'],
                   color=['#3498db', '#2ecc71', '#9b59b6']) # Assign colors
    plt.ylabel('Number of Studies')
    plt.xlabel('Methodology Type')
    plt.title('Research Methodologies Used', fontsize=16, fontweight='bold')
    plt.xticks(rotation=0) # Horizontal labels

    # Add value labels
    for bar in bars:
        height = bar.get_height()
        if height > 0:
            plt.text(bar.get_x() + bar.get_width()/2., height,
                     f'{int(height)}', ha='center', va='bottom', fontweight='bold', fontsize=11)

    plt.tight_layout()
    plt.savefig(FIGURES_DIR / 'metodologias_utilizadas.png', dpi=300, bbox_inches='tight')
    logging.info(f"Saved methodology distribution plot to {FIGURES_DIR / 'metodologias_utilizadas.png'}")
    plt.show()
else:
    logging.warning("No valid methodology data found or columns missing. Skipping methodology plot.")


# %% [markdown]
# ### 7. Intervention Type Analysis

# %%
# --- Intervention Type Analysis ---
if 'Intervention' in df.columns:
    # Define intervention categories and keywords
    intervention_keywords = {
        'AR/VR': ['ar', 'vr', 'augment', 'virtual', 'immersive'],
        'Games/Gamification': ['game', 'gamif', 'serious game'],
        'Educational Software': ['software', 'platform', 'application', 'tool', 'system'], # Broader category
        'Adaptive System': ['adapt', 'personalized', 'intelligent tutor'],
        'Robotics': ['robot', 'robotics'],
        # Add more categories as needed
    }
    # Initialize counts
    intervention_counts = {key: 0 for key in intervention_keywords.keys()}
    intervention_counts['Other'] = 0 # Catch-all category

    # Classify interventions
    for intervention in df['Intervention'].dropna():
        intervention_lower = intervention.lower()
        classified = False
        for type_name, keywords in intervention_keywords.items():
            if any(term in intervention_lower for term in keywords):
                intervention_counts[type_name] += 1
                classified = True
                break # Assign to the first matching category
        if not classified:
            intervention_counts['Other'] += 1

    # Filter out categories with zero counts for plotting
    intervention_counts_plot = {k: v for k, v in intervention_counts.items() if v > 0}

    if intervention_counts_plot:
        logging.info("Generating intervention types plot (Plotly)...")
        # Create Plotly bar chart
        fig = go.Figure(data=[
            go.Bar(
                x=list(intervention_counts_plot.keys()),
                y=list(intervention_counts_plot.values()),
                text=list(intervention_counts_plot.values()),
                textposition='auto',
                marker=dict(color=px.colors.qualitative.Pastel) # Use a Plotly color sequence
            )
        ])
        fig.update_layout(
            title='Types of Digital Interventions Used',
            xaxis_title='Intervention Type',
            yaxis_title='Number of Studies',
            xaxis_tickangle=-45, # Angle labels if needed
            showlegend=False,
            template='plotly_white',
            height=600 # Adjust height if needed
        )
        # Save as HTML and show
        html_path = FIGURES_DIR / 'tipos_intervencion.html'
        fig.write_html(html_path)
        logging.info(f"Saved intervention types plot to {html_path}")
        fig.show()
    else:
        logging.warning("No valid intervention data found for intervention plot.")
else:
    logging.warning("Column 'Intervention' not found. Skipping intervention analysis.")

# %% [markdown]
# ### 8. Evaluation Instrument Analysis

# %%
# --- Evaluation Instrument Analysis ---
instrument_cols = {
    'Cognitive': 'Cognitive Instruments',
    'Affective': 'Affective Instruments',
    'Other': 'Other Instruments'
}
instruments_analysis = {}
valid_instrument_cols_found = False

for instr_name, col_name in instrument_cols.items():
    if col_name in df.columns:
        instruments_analysis[instr_name] = df[col_name].notna().sum()
        valid_instrument_cols_found = True
        logging.info(f"Column '{col_name}' found. Count for {instr_name}: {instruments_analysis[instr_name]}")
    else:
        instruments_analysis[instr_name] = 0 # Assign 0 if column is missing
        logging.warning(f"Instrument column '{col_name}' not found.")

# Filter out categories with zero counts
instruments_analysis_plot = {k:v for k,v in instruments_analysis.items() if v > 0}

if valid_instrument_cols_found and instruments_analysis_plot:
    logging.info("Generating evaluation instruments radar plot...")
    # Create Radar Chart
    categories = list(instruments_analysis_plot.keys())
    values = list(instruments_analysis_plot.values())
    num_vars = len(categories)

    # Compute angle for each axis
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
    # The plot is circular, so we need to "complete the loop"
    values += values[:1]
    angles += angles[:1]

    fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))
    # Draw one axe per variable + add labels
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories)
    # Draw ylabels
    ax.set_yticks(np.arange(0, max(values) + 5, step=max(1, int((max(values)+5)/5)))) # Dynamic y-ticks
    ax.set_ylim(0, max(values) + 5)

    # Plot data
    ax.plot(angles, values, linewidth=2, linestyle='solid', color='darkviolet', marker='o')
    # Fill area
    ax.fill(angles, values, 'darkviolet', alpha=0.3)

    ax.set_title('Evaluation Instruments Used', fontsize=16, fontweight='bold', pad=20)

    # Add value labels (optional, can clutter radar charts)
    # for angle, value in zip(angles[:-1], values[:-1]):
    #     ax.text(angle, value + 1, str(value), ha='center', va='center', fontweight='bold', fontsize=10)

    plt.tight_layout() # Adjust layout
    plt.savefig(FIGURES_DIR / 'instrumentos_evaluacion.png', dpi=300, bbox_inches='tight')
    logging.info(f"Saved evaluation instruments plot to {FIGURES_DIR / 'instrumentos_evaluacion.png'}")
    plt.show()
else:
    logging.warning("No valid instrument data found or columns missing. Skipping instrument plot.")

# %% [markdown]
# ### 9. General Results Analysis (NLP)

# %%
# --- NLP Analysis of General Results ---
if 'General Results' in df.columns and df['General Results'].notna().any():
    logging.info("Performing NLP analysis on 'General Results'...")
    all_results_text = ' '.join(df['General Results'].dropna().astype(str).tolist())

    # Tokenization and cleaning
    try:
        tokens = word_tokenize(all_results_text.lower())
        # Enhanced filtering: include numbers if relevant, handle hyphens better?
        # For now, keep simple alpha filtering.
        filtered_tokens = [token for token in tokens if token.isalpha() and token not in stop_words and len(token) > 2] # Min length 3

        if filtered_tokens:
            # Word frequency
            word_freq = Counter(filtered_tokens)
            top_N_words = 30 # Number of top words to show
            top_words_dict = dict(word_freq.most_common(top_N_words))

            # --- Create Word Cloud ---
            logging.info("Generating Word Cloud...")
            plt.figure(figsize=(15, 8)) # Adjusted size
            wordcloud = WordCloud(width=1200, height=600, # Adjusted size
                                  background_color='white',
                                  colormap='viridis', # Example colormap
                                  max_words=100, # Max words in cloud
                                  contour_width=1, contour_color='steelblue', # Added contour
                                  collocations=False # Avoid bigrams for simplicity
                                  ).generate_from_frequencies(word_freq)
            plt.imshow(wordcloud, interpolation='bilinear')
            plt.axis('off')
            plt.title('Word Cloud of General Results', fontsize=20, fontweight='bold')
            plt.tight_layout(pad=0)
            plt.savefig(FIGURES_DIR / 'wordcloud_resultados.png', dpi=300, bbox_inches='tight')
            logging.info(f"Saved word cloud to {FIGURES_DIR / 'wordcloud_resultados.png'}")
            plt.show()

            # --- Create Bar Chart for Top Words ---
            if top_words_dict:
                logging.info("Generating Top Words bar chart...")
                plt.figure(figsize=(12, 10)) # Adjusted size for vertical space
                words_df = pd.DataFrame.from_dict(top_words_dict, orient='index', columns=['Frequency'])
                words_df = words_df.sort_values('Frequency', ascending=True) # Sort for horizontal bar chart
                bars = plt.barh(words_df.index, words_df['Frequency'], color='cornflowerblue') # Horizontal bar chart
                plt.xlabel('Frequency')
                plt.ylabel('Keyword')
                plt.title(f'Top {top_N_words} Keywords in General Results', fontsize=16, fontweight='bold')

                # Add labels
                for bar in bars:
                    width = bar.get_width()
                    plt.text(width + 0.1, bar.get_y() + bar.get_height()/2., f'{int(width)}',
                             ha='left', va='center', fontsize=9)

                plt.tight_layout()
                plt.savefig(FIGURES_DIR / 'top_palabras_clave.png', dpi=300, bbox_inches='tight')
                logging.info(f"Saved top keywords plot to {FIGURES_DIR / 'top_palabras_clave.png'}")
                plt.show()
            else:
                logging.warning("No top words found after frequency calculation.")
        else:
            logging.warning("'General Results' contained no valid tokens after filtering.")
    except Exception as e:
        logging.error(f"An error occurred during NLP processing: {e}")
else:
    logging.warning("Column 'General Results' not found or empty. Skipping NLP analysis.")

# %% [markdown]
# ### 10. Thematic Analysis (Based on Objectives/Keywords)

# %%
# --- Thematic Analysis ---
if 'General Results' in df.columns and 'word_freq' in locals() and word_freq: # Check if word_freq exists from previous step
    logging.info("Performing thematic analysis...")
    # Define themes based on the review's objectives or common topics
    # (These keywords should be reviewed and adapted based on the specific context)
    themes = {
        'Motivation/Engagement': ['motivation', 'motivate', 'motivated', 'engagement', 'engaged', 'enjoyment', 'enjoy', 'interest', 'fun', 'attitude'],
        'Performance/Achievement': ['performance', 'achievement', 'improvement', 'improved', 'score', 'scores', 'results', 'better', 'learning', 'gain', 'knowledge', 'grades'],
        'Skills/Understanding': ['skills', 'skill', 'ability', 'abilities', 'understanding', 'comprehension', 'fluency', 'competence', 'reasoning', 'thinking'],
        'Affective/Anxiety': ['anxiety', 'stress', 'confidence', 'self-efficacy', 'emotion', 'emotional', 'affective', 'attitude'], # Attitude might overlap
        'Cognitive Load/Usability': ['cognitive', 'load', 'usability', 'ease', 'interaction', 'interface', 'difficulty'],
        'Technology Specific': ['digital', 'technology', 'software', 'game', 'games', 'interactive', 'virtual', 'ar', 'vr', 'robot', 'platform']
    }

    # Count keyword occurrences for each theme
    theme_counts = {}
    for theme, keywords in themes.items():
        # Sum frequencies from the word_freq Counter object
        count = sum(word_freq.get(keyword, 0) for keyword in keywords)
        theme_counts[theme] = count

    # Filter out themes with zero counts
    theme_counts_plot = {k:v for k,v in theme_counts.items() if v > 0}

    if theme_counts_plot:
        logging.info("Generating thematic analysis plot...")
        # Visualize thematic analysis
        plt.figure(figsize=(12, 8))
        themes_df = pd.DataFrame.from_dict(theme_counts_plot, orient='index', columns=['Frequency'])
        themes_df = themes_df.sort_values('Frequency', ascending=True) # Sort for horizontal bar chart
        colors = plt.cm.viridis(np.linspace(0, 1, len(themes_df))) # Use a colormap
        bars = plt.barh(themes_df.index, themes_df['Frequency'], color=colors)
        plt.xlabel('Total Keyword Frequency')
        plt.ylabel('Theme')
        plt.title('Thematic Analysis Based on Keywords in Results', fontsize=16, fontweight='bold')

        # Add labels
        for bar in bars:
            width = bar.get_width()
            plt.text(width + 0.1, bar.get_y() + bar.get_height()/2., f'{int(width)}',
                     ha='left', va='center', fontsize=10)

        plt.tight_layout()
        plt.savefig(FIGURES_DIR / 'analisis_tematico.png', dpi=300, bbox_inches='tight')
        logging.info(f"Saved thematic analysis plot to {FIGURES_DIR / 'analisis_tematico.png'}")
        plt.show()
    else:
         logging.warning("No theme keywords found in the 'General Results' text.")
else:
    logging.warning("Skipping thematic analysis because 'General Results' data or word frequencies are unavailable.")


# %% [markdown]
# ### 11. Correlation Analysis (Example: Methodology vs. Intervention)

# %%
# --- Correlation Analysis ---
# Example: Chi-squared test between Mixed Methods use and AR/VR intervention use
# Requires columns 'Mixed Methods' and 'Intervention'
if 'Mixed Methods' in df.columns and 'Intervention' in df.columns:
    logging.info("Performing Chi-squared test (Mixed Methods vs AR/VR)...")
    # Create a contingency table
    # Define a helper function to classify intervention for this specific analysis
    def is_arvr(intervention_str):
        if pd.isna(intervention_str): return False
        return any(term in str(intervention_str).lower() for term in ['ar', 'vr', 'augment', 'virtual'])

    contingency_table = pd.crosstab(
        df['Mixed Methods'].notna(), # Boolean: True if Mixed Methods used
        df['Intervention'].apply(is_arvr) # Boolean: True if AR/VR used
    )
    # Rename for clarity
    contingency_table.index = ['Not Mixed', 'Mixed']
    contingency_table.columns = ['Not AR/VR', 'AR/VR']

    if contingency_table.size > 0 and contingency_table.sum().sum() > 0:
        try:
            chi2, p_value, dof, expected = chi2_contingency(contingency_table)
            print(f"\nContingency Table (Mixed Methods vs AR/VR):")
            print(contingency_table)
            print(f"\nChi-squared Test:  = {chi2:.2f}, p-value = {p_value:.4f}, DoF = {dof}")
            if p_value < 0.05:
                print("Result: Statistically significant association found (p < 0.05).")
            else:
                print("Result: No statistically significant association found (p >= 0.05).")

            # Visualize the contingency table as a heatmap
            logging.info("Generating heatmap for correlation...")
            plt.figure(figsize=(8, 6)) # Adjusted size
            sns.heatmap(contingency_table, annot=True, fmt='d', cmap='viridis', # Changed colormap
                        linewidths=.5, linecolor='black', # Added lines
                        cbar_kws={'label': 'Number of Studies'})
            plt.title('Association: Mixed Methods vs AR/VR Intervention', fontsize=14, fontweight='bold') # Adjusted size
            plt.xlabel('AR/VR Intervention Used')
            plt.ylabel('Mixed Methods Used')
            plt.tight_layout()
            plt.savefig(FIGURES_DIR / 'correlacion_metodo_intervencion.png', dpi=300, bbox_inches='tight')
            logging.info(f"Saved correlation heatmap to {FIGURES_DIR / 'correlacion_metodo_intervencion.png'}")
            plt.show()

        except ValueError as ve:
             logging.warning(f"Could not perform Chi-squared test (likely insufficient data variation): {ve}")
        except Exception as e:
            logging.error(f"An error occurred during Chi-squared test or heatmap generation: {e}")
    else:
         logging.warning("Contingency table is empty or contains no data. Skipping Chi-squared test and heatmap.")
else:
    logging.warning("Required columns ('Mixed Methods', 'Intervention') not found. Skipping correlation analysis.")


# %% [markdown]
# ### 12. Relationship Network Analysis

# %%
# --- Network Analysis ---
# Example: Network relating Countries, Methodologies, and Interventions
# Requires 'FirstCountry', method columns ('Quantitative Methods', etc.), 'Intervention'
required_net_cols = ['FirstCountry', 'Quantitative Methods', 'Qualitative Methods', 'Mixed Methods', 'Intervention']

if all(col in df.columns for col in required_net_cols):
    logging.info("Generating relationship network graph...")
    G = nx.Graph() # Undirected graph for co-occurrence

    # Define node types and colors
    node_colors_map = {'country': 'skyblue', 'intervention': 'lightcoral', 'method': 'lightgreen'}

    # Add Country nodes
    countries = df['FirstCountry'].dropna().unique()
    for country in countries:
        G.add_node(country, node_type='country')

    # Add Intervention nodes (simplified based on previous classification)
    # Re-use intervention_keywords dictionary if available from Step 7
    if 'intervention_keywords' in locals():
        intervention_types_unique = list(intervention_keywords.keys())
        for intervention in intervention_types_unique:
             G.add_node(intervention, node_type='intervention')
        G.add_node('Other Intervention', node_type='intervention') # Add catch-all if needed
    else:
        logging.warning("Intervention keywords not defined; using hardcoded list for network.")
        intervention_types_unique = ['AR/VR', 'Games/Gamification', 'Educational Software', 'Adaptive System', 'Robotics', 'Other Intervention']
        for intervention in intervention_types_unique:
             G.add_node(intervention, node_type='intervention')


    # Add Method nodes
    method_types = ['Quantitative', 'Qualitative', 'Mixed']
    for method in method_types:
        G.add_node(method, node_type='method')

    # Add edges based on co-occurrence in studies
    for _, row in df.iterrows():
        country = row['FirstCountry']
        if pd.isna(country): continue # Skip rows without a country

        # Link Country to Intervention Type
        intervention_classified = False
        if pd.notna(row['Intervention']) and 'intervention_keywords' in locals():
            intervention_lower = row['Intervention'].lower()
            for type_name, keywords in intervention_keywords.items():
                if any(term in intervention_lower for term in keywords):
                    if G.has_node(country) and G.has_node(type_name): # Check nodes exist
                        G.add_edge(country, type_name)
                        intervention_classified = True
                        break
            if not intervention_classified and G.has_node(country) and G.has_node('Other Intervention'):
                 G.add_edge(country, 'Other Intervention')


        # Link Country to Methodology
        if pd.notna(row['Quantitative Methods']) and G.has_node(country) and G.has_node('Quantitative'):
            G.add_edge(country, 'Quantitative')
        if pd.notna(row['Qualitative Methods']) and G.has_node(country) and G.has_node('Qualitative'):
            G.add_edge(country, 'Qualitative')
        if pd.notna(row['Mixed Methods']) and G.has_node(country) and G.has_node('Mixed'):
            G.add_edge(country, 'Mixed')

        # Optional: Link Intervention to Methodology (more complex graph)
        # ... (add logic here if desired) ...

    # Remove isolated nodes (nodes with no connections) for cleaner visualization
    isolated = list(nx.isolates(G))
    G.remove_nodes_from(isolated)
    logging.info(f"Removed {len(isolated)} isolated nodes.")

    if G.number_of_nodes() > 0 and G.number_of_edges() > 0:
        # Visualize the network
        plt.figure(figsize=(22, 22)) # Larger figure for complex networks
        pos = nx.spring_layout(G, k=0.5, iterations=50, seed=42) # Adjust k for spacing

        # Get node colors based on type attribute
        node_colors = [node_colors_map.get(G.nodes[node].get('node_type', 'unknown'), 'grey') for node in G.nodes()]

        # Draw nodes (adjust size based on degree?)
        degrees = dict(G.degree())
        node_sizes = [degrees.get(node, 1) * 100 + 500 for node in G.nodes()] # Size by degree

        nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=node_sizes, alpha=0.8)
        nx.draw_networkx_edges(G, pos, alpha=0.3, edge_color='grey', width=0.5) # Thinner edges
        nx.draw_networkx_labels(G, pos, font_size=9, font_weight='normal') # Smaller labels

        plt.title('Co-occurrence Network: Countries, Methodologies & Interventions', fontsize=20, fontweight='bold')
        plt.axis('off')
        plt.tight_layout()
        plt.savefig(FIGURES_DIR / 'red_relaciones.png', dpi=300, bbox_inches='tight')
        logging.info(f"Saved relationship network plot to {FIGURES_DIR / 'red_relaciones.png'}")
        plt.show()
    else:
         logging.warning("Network graph has no nodes or edges after processing. Skipping visualization.")

else:
    missing_cols = [col for col in required_net_cols if col not in df.columns]
    logging.warning(f"Required columns missing for network analysis: {missing_cols}. Skipping network graph.")

# %% [markdown]
# ### 13. Sentiment Analysis of Results

# %%
# --- Sentiment Analysis ---
# Simple keyword-based sentiment analysis
if 'General Results' in df.columns:
    logging.info("Performing simple sentiment analysis on 'General Results'...")
    def analyze_sentiment(text):
        if pd.isna(text): return 'Neutral'
        text_lower = str(text).lower()
        # Expanded keyword lists
        positive_words = ['improved', 'significant', 'positive', 'better', 'increased', 'effective',
                          'successful', 'enhanced', 'beneficial', 'well', 'high', 'good', 'promising', 'motivated', 'engaged']
        negative_words = ['loss', 'lower', 'negative', 'decreased', 'reduced', 'worse', 'difficult',
                          'challenge', 'problem', 'issue', 'no difference', 'no significant', 'limited']

        pos_count = sum(word in text_lower for word in positive_words)
        neg_count = sum(word in text_lower for word in negative_words)

        if pos_count > neg_count: return 'Positive'
        if neg_count > pos_count: return 'Negative'
        return 'Neutral'

    df['Sentiment'] = df['General Results'].apply(analyze_sentiment)
    sentiment_distribution = df['Sentiment'].value_counts()
    print("\nSentiment Distribution:")
    print(sentiment_distribution)

    # --- Sentiment by Intervention Type ---
    if 'Intervention' in df.columns and 'intervention_keywords' in locals():
        logging.info("Analyzing sentiment distribution by intervention type...")
        # Create DataFrame linking Intervention Type and Sentiment
        sentiment_intervention_list = []
        for _, row in df.iterrows():
            intervention_type = 'Other Intervention' # Default
            classified = False
            if pd.notna(row['Intervention']):
                intervention_lower = row['Intervention'].lower()
                for type_name, keywords in intervention_keywords.items():
                    if any(term in intervention_lower for term in keywords):
                        intervention_type = type_name
                        classified = True
                        break
            sentiment_intervention_list.append({
                'Intervention': intervention_type,
                'Sentiment': row['Sentiment']
            })

        sentiment_by_intervention_df = pd.DataFrame(sentiment_intervention_list)

        # Create stacked bar chart
        sentiment_counts = sentiment_by_intervention_df.groupby(['Intervention', 'Sentiment']).size().unstack(fill_value=0)
        # Ensure standard sentiment order and colors
        sentiment_order = ['Positive', 'Neutral', 'Negative']
        sentiment_colors = {'Positive': '#2ecc71', 'Neutral': '#95a5a6', 'Negative': '#e74c3c'}
        # Reindex columns to ensure order and include missing sentiments with 0 count
        sentiment_counts = sentiment_counts.reindex(columns=sentiment_order, fill_value=0)

        if not sentiment_counts.empty:
            logging.info("Generating sentiment by intervention plot...")
            ax = sentiment_counts.plot(kind='bar', stacked=True, figsize=(12, 8),
                                       color=[sentiment_colors[col] for col in sentiment_counts.columns]) # Use defined colors

            plt.xlabel('Intervention Type')
            plt.ylabel('Number of Studies')
            plt.title('Sentiment Analysis of Results by Intervention Type', fontsize=16, fontweight='bold')
            plt.legend(title='Sentiment', bbox_to_anchor=(1.05, 1), loc='upper left') # Legend outside plot
            plt.xticks(rotation=45, ha='right')
            plt.tight_layout(rect=[0, 0, 0.85, 1]) # Adjust layout for external legend

            # Add total count labels on top of bars
            for i, total in enumerate(sentiment_counts.sum(axis=1)):
                if total > 0:
                    ax.text(i, total + 0.5, f'n={total}', ha='center', va='bottom', fontsize=9)


            plt.savefig(FIGURES_DIR / 'sentimiento_por_intervencion.png', dpi=300, bbox_inches='tight')
            logging.info(f"Saved sentiment by intervention plot to {FIGURES_DIR / 'sentimiento_por_intervencion.png'}")
            plt.show()
        else:
             logging.warning("Could not generate sentiment counts by intervention.")
    else:
        logging.warning("Skipping sentiment by intervention analysis due to missing columns or prior errors.")
else:
    logging.warning("Column 'General Results' not found. Skipping sentiment analysis.")


# %% [markdown]
# ### 14. Interactive Dashboard (Plotly)

# %%
# --- Interactive Dashboard ---
# Check if necessary data for dashboard exists
dashboard_possible = True
# required_data_keys = ['country_counts', 'year_counts', 'methods_analysis', 'intervention_counts_plot', 'instruments_analysis_plot', 'theme_counts_plot']
# for key in required_data_keys:
#     if key not in locals() or not locals()[key]:
#         logging.warning(f"Data for dashboard component '{key}' is missing or empty. Dashboard may be incomplete.")
#         # Decide if dashboard generation should be skipped entirely
#         # dashboard_possible = False
#         # break # Or just skip the specific trace

if dashboard_possible: # This flag might not even be needed now
    logging.info("Generating interactive dashboard (Plotly)...")
    # Create subplots (same as before)
    fig = make_subplots(
        rows=3, cols=2,
        subplot_titles=('Top 10 Countries', 'Temporal Trend',
                        'Methodologies Used', 'Intervention Types',
                        'Evaluation Instruments', 'Thematic Analysis (Keywords)'),
        specs=[[{"type": "bar"}, {"type": "scatter"}],
               [{"type": "pie"}, {"type": "bar"}],
               [{"type": "bar"}, {"type": "bar"}]]
    )

    # Trace 1: Country Distribution (Top 10)
    # This check correctly handles the Series
    if 'country_counts' in locals() and isinstance(locals()['country_counts'], pd.Series) and not locals()['country_counts'].empty:
        top_10_countries = country_counts.head(10)
        fig.add_trace(go.Bar(x=top_10_countries.index, y=top_10_countries.values, name='Countries', marker_color='rgb(102,194,165)'), row=1, col=1)
    else:
        logging.warning("Skipping dashboard country plot due to missing/empty data.") # Keep individual warnings

    # Trace 2: Temporal Trend
    # This check correctly handles the Series
    if 'year_counts' in locals() and isinstance(locals()['year_counts'], pd.Series) and not locals()['year_counts'].empty:
        fig.add_trace(go.Scatter(x=year_counts.index, y=year_counts.values, mode='lines+markers', name='Years', line=dict(color='rgb(252,141,98)')), row=1, col=2)
    else:
        logging.warning("Skipping dashboard temporal plot due to missing/empty data.")

    # Trace 3: Methodologies (Pie)
    # This check correctly handles the dictionary
    if 'methods_analysis' in locals() and isinstance(locals()['methods_analysis'], dict) and locals()['methods_analysis']:
         # Exclude methods with 0 count if any remain
        methods_analysis_plot = {k: v for k, v in methods_analysis.items() if v > 0}
        if methods_analysis_plot: # Check if dict is not empty after filtering
            labels = list(methods_analysis_plot.keys())
            values = list(methods_analysis_plot.values())
            fig.add_trace(go.Pie(labels=labels, values=values, name='Methodologies', hole=0.3, marker_colors=px.colors.sequential.Blues_r), row=2, col=1)
        else:
            logging.warning("Skipping dashboard methodology plot - no data after filtering zeros.")
    else:
        logging.warning("Skipping dashboard methodology plot due to missing/empty data.")

    # Trace 4: Intervention Types
    # This check correctly handles the dictionary
    if 'intervention_counts_plot' in locals() and isinstance(locals()['intervention_counts_plot'], dict) and locals()['intervention_counts_plot']:
        labels = list(intervention_counts_plot.keys())
        values = list(intervention_counts_plot.values())
        if labels: # Check if dict is not empty
            fig.add_trace(go.Bar(x=labels, y=values, name='Interventions', marker_color='rgb(141,160,203)'), row=2, col=2)
    else:
        logging.warning("Skipping dashboard intervention plot due to missing/empty data.")

    # Trace 5: Evaluation Instruments (Bar)
    # This check correctly handles the dictionary
    if 'instruments_analysis_plot' in locals() and isinstance(locals()['instruments_analysis_plot'], dict) and locals()['instruments_analysis_plot']:
        labels = list(instruments_analysis_plot.keys())
        values = list(instruments_analysis_plot.values())
        if labels: # Check if dict is not empty
            fig.add_trace(go.Bar(x=labels, y=values, name='Instruments', marker_color='rgb(231,138,195)'), row=3, col=1)
    else:
        logging.warning("Skipping dashboard instruments plot due to missing/empty data.")

    # Trace 6: Thematic Analysis
    # This check correctly handles the dictionary
    if 'theme_counts_plot' in locals() and isinstance(locals()['theme_counts_plot'], dict) and locals()['theme_counts_plot']:
        labels = list(theme_counts_plot.keys())
        values = list(theme_counts_plot.values())
        if labels: # Check if dict is not empty
            fig.add_trace(go.Bar(x=labels, y=values, name='Themes', marker_color='rgb(166,216,84)'), row=3, col=2)
    else:
        logging.warning("Skipping dashboard thematic plot due to missing/empty data.")


    # Update layout (same as before)
    fig.update_layout(height=1200, showlegend=False,
                      title_text="Interactive Dashboard - Systematic Review Analysis",
                      template='plotly_white'
                      )
    # Save and show (same as before)
    dashboard_path = FIGURES_DIR / 'dashboard_interactivo.html'
    fig.write_html(dashboard_path)
    logging.info(f"Saved interactive dashboard to {dashboard_path}")
    fig.show()



# %% [markdown]
# ### 15. Synthesis Visualizations

# %%
# --- Synthesis Plot ---
logging.info("Attempting to generate synthesis plot...")

# Recalculate or ensure necessary variables are available just before plotting

# Recalculate year_intervention_df (example with AR/VR and Games)
# Ensure 'intervention_keywords' exists from previous steps
year_intervention_df = pd.DataFrame(columns=['Year', 'Type', 'Count']) # Initialize empty
if ('Year' in df.columns and 'Intervention' in df.columns and
    'intervention_keywords' in locals() and isinstance(locals()['intervention_keywords'], dict)):
    year_intervention_list = []
    # ... (rest of the calculation for year_intervention_df as in the previous code) ...
    for year in df['Year'].dropna().unique():
        year_data = df[df['Year'] == year]
        counts = {'AR/VR': 0, 'Games/Gamification': 0} # Initialize for the year
        for _, row in year_data.iterrows():
            if pd.notna(row['Intervention']):
                intervention_lower = row['Intervention'].lower()
                # Check against AR/VR keywords
                if any(term in intervention_lower for term in intervention_keywords.get('AR/VR', [])):
                    counts['AR/VR'] += 1
                # Check against Games keywords (use the correct key)
                elif any(term in intervention_lower for term in intervention_keywords.get('Games/Gamification', [])):
                     counts['Games/Gamification'] += 1
        # Append non-zero counts
        if counts['AR/VR'] > 0: year_intervention_list.append({'Year': year, 'Type': 'AR/VR', 'Count': counts['AR/VR']})
        if counts['Games/Gamification'] > 0: year_intervention_list.append({'Year': year, 'Type': 'Games/Gamification', 'Count': counts['Games/Gamification']})

    if year_intervention_list:
        year_intervention_df = pd.DataFrame(year_intervention_list)
        logging.info("Successfully recalculated year_intervention_df for synthesis plot.")
    else:
        logging.warning("No data found to populate year_intervention_df for synthesis plot.")
else:
    logging.warning("Cannot create year-intervention data for synthesis plot due to missing columns or keywords.")


# Recalculate effectiveness_rate
effectiveness_rate = pd.Series(dtype=float) # Initialize empty
# Ensure 'sentiment_by_intervention_df' exists from previous steps
if 'sentiment_by_intervention_df' in locals() and isinstance(locals()['sentiment_by_intervention_df'], pd.DataFrame) and not locals()['sentiment_by_intervention_df'].empty:
     effectiveness_data = sentiment_by_intervention_df[sentiment_by_intervention_df['Sentiment'] == 'Positive'].groupby('Intervention').size()
     total_by_intervention = sentiment_by_intervention_df.groupby('Intervention').size()
     # Avoid division by zero if an intervention type has 0 total studies
     effectiveness_rate = (effectiveness_data.div(total_by_intervention, fill_value=0) * 100).fillna(0)
     logging.info("Successfully recalculated effectiveness_rate for synthesis plot.")
else:
    logging.warning("Cannot create effectiveness data for synthesis plot due to missing prerequisite data.")


# ---- REMOVE THE PROBLEMATIC PRE-CHECK LOOP ----
# synthesis_possible = True
# required_synth_keys = ['country_counts', 'year_intervention_df', 'theme_counts_plot', 'effectiveness_rate'] # Adjusted keys slightly
# for key in required_synth_keys:
#      if key not in locals() or not isinstance(locals()[key], (pd.DataFrame, pd.Series, dict)) or pd.Series(locals()[key]).empty: # <--- ERROR HERE
#          logging.warning(f"Data for synthesis component '{key}' is missing or empty.")
#          synthesis_possible = False
# ---- END REMOVAL ----


# Proceed to plotting, individual checks within the plotting code will handle missing data
logging.info("Generating synthesis plot panels...")
fig, axes = plt.subplots(2, 2, figsize=(20, 16))
ax1, ax2, ax3, ax4 = axes.flatten() # Flatten axes array

# Panel 1: Top 10 Countries
# Check if country_counts exists and is a non-empty Series
if 'country_counts' in locals() and isinstance(locals()['country_counts'], pd.Series) and not locals()['country_counts'].empty:
    top_countries = country_counts.head(10)
    ax1.bar(top_countries.index, top_countries.values, color='skyblue')
    ax1.set_title('Top 10 Countries by Study Count', fontsize=14, fontweight='bold')
    ax1.tick_params(axis='x', labelrotation=45, labelsize=10) # Use ha='right'
    ax1.set_ylabel('Number of Studies')
    for i, v in enumerate(top_countries.values): ax1.text(i, v + 0.1, str(v), ha='center', va='bottom', fontsize=9)
else:
    logging.warning("Synthesis Panel 1: Country data missing.")
    ax1.set_title("Country Data Missing")
    ax1.axis('off')

# Panel 2: Intervention Evolution (Stacked Bar)
# Check if year_intervention_df exists and is a non-empty DataFrame
if 'year_intervention_df' in locals() and isinstance(locals()['year_intervention_df'], pd.DataFrame) and not locals()['year_intervention_df'].empty:
    # Pivot requires numeric index if year isn't already the index
    if 'Year' in year_intervention_df.columns:
        year_intervention_df.set_index('Year', inplace=True)
    # Check if necessary columns exist after potential index setting
    if 'Type' in year_intervention_df.columns and 'Count' in year_intervention_df.columns:
         pivot_df = year_intervention_df.pivot(columns='Type', values='Count').fillna(0)
         if not pivot_df.empty:
              # Ensure specific columns exist for plotting if needed, e.g., AR/VR and Games
              cols_to_plot = [col for col in ['AR/VR', 'Games/Gamification'] if col in pivot_df.columns]
              if cols_to_plot:
                  pivot_df[cols_to_plot].plot(kind='bar', stacked=True, ax=ax2, color=['#e74c3c', '#3498db'], width=0.8)
                  ax2.set_title('Evolution of Key Interventions', fontsize=14, fontweight='bold')
                  ax2.set_xlabel('Year')
                  ax2.set_ylabel('Number of Studies')
                  ax2.legend(title='Intervention Type', fontsize=10)
                  ax2.tick_params(axis='x', labelrotation=0)
              else:
                  logging.warning("Synthesis Panel 2: Relevant intervention types (AR/VR, Games) not found in pivot.")
                  ax2.set_title("Key Intervention Data Missing"); ax2.axis('off')

         else:
            logging.warning("Synthesis Panel 2: Pivot table is empty.")
            ax2.set_title("Intervention Evolution Data Missing"); ax2.axis('off')
    else:
        logging.warning("Synthesis Panel 2: 'Type' or 'Count' column missing in year_intervention_df.")
        ax2.set_title("Intervention Evolution Data Missing"); ax2.axis('off')
else:
    logging.warning("Synthesis Panel 2: Intervention evolution data missing.")
    ax2.set_title("Intervention Evolution Data Missing"); ax2.axis('off')

# Panel 3: Thematic Analysis (Top Themes)
# Check if theme_counts_plot exists and is a non-empty dict
if 'theme_counts_plot' in locals() and isinstance(locals()['theme_counts_plot'], dict) and locals()['theme_counts_plot']:
    themes_sorted = pd.Series(theme_counts_plot).sort_values(ascending=False).head(10) # Top 10 themes
    colors = plt.cm.Paired(np.linspace(0, 1, len(themes_sorted)))
    ax3.bar(themes_sorted.index, themes_sorted.values, color=colors)
    ax3.set_title('Top Themes from Results', fontsize=14, fontweight='bold')
    ax3.set_ylabel('Keyword Frequency')
    ax3.tick_params(axis='x', labelrotation=45, labelsize=10)
else:
    logging.warning("Synthesis Panel 3: Thematic data missing.")
    ax3.set_title("Thematic Data Missing"); ax3.axis('off')

# Panel 4: Effectiveness Rate by Intervention
# Check if effectiveness_rate exists and is a non-empty Series
if 'effectiveness_rate' in locals() and isinstance(locals()['effectiveness_rate'], pd.Series) and not locals()['effectiveness_rate'].empty:
    effectiveness_rate = effectiveness_rate.sort_values(ascending=False)
    ax4.bar(effectiveness_rate.index, effectiveness_rate.values,
            color='mediumseagreen') # Use a green color
    ax4.set_title('Success Rate by Intervention Type', fontsize=14, fontweight='bold')
    ax4.set_ylabel('Positive Results (%)')
    ax4.set_ylim(0, 105) # Slightly above 100%
    ax4.tick_params(axis='x', labelrotation=45, labelsize=10)
    # Add percentage labels
    for i, v in enumerate(effectiveness_rate.values):
        ax4.text(i, v + 1, f'{v:.1f}%', ha='center', va='bottom', fontsize=10, fontweight='bold')
else:
    logging.warning("Synthesis Panel 4: Effectiveness data missing.")
    ax4.set_title("Effectiveness Data Missing"); ax4.axis('off')


# Final plot adjustments and saving
plt.suptitle("Synthesis of Key Findings", fontsize=20, fontweight='bold', y=1.02)
plt.tight_layout(rect=[0, 0.03, 1, 0.98])
plt.savefig(FIGURES_DIR / 'sintesis_visual.png', dpi=300, bbox_inches='tight')
logging.info(f"Saved synthesis plot to {FIGURES_DIR / 'sintesis_visual.png'}")
plt.show()

# %% [markdown]
# ### 16. Co-occurrence Analysis of Terms

# %%
# --- Term Co-occurrence Analysis ---
if 'General Results' in df.columns and 'word_freq' in locals() and word_freq:
    logging.info("Performing term co-occurrence analysis...")
    from itertools import combinations

    cooccurrence_dict = defaultdict(int) # Use defaultdict
    window_size = 5 # Window size for co-occurrence

    for result_text in df['General Results'].dropna().astype(str):
        try:
            tokens = word_tokenize(result_text.lower())
            filtered = [token for token in tokens if token.isalpha() and token not in stop_words and len(token) > 2]

            # Generate combinations within the window
            for i in range(len(filtered) - 1):
                # Limit window to avoid excessive pairs
                window_end = min(i + window_size, len(filtered))
                # Create pairs within the window
                for j in range(i + 1, window_end):
                    # Ensure words are different and sort for consistent pairing
                    if filtered[i] != filtered[j]:
                         pair = tuple(sorted((filtered[i], filtered[j])))
                         cooccurrence_dict[pair] += 1
        except Exception as e:
            logging.warning(f"Skipping a result text due to tokenization/processing error: {e}")


    if cooccurrence_dict:
        # Get top N co-occurring pairs
        top_N_pairs = 20
        # Filter pairs containing very common words if needed (e.g., 'students', 'learning')
        # common_filter = {'students', 'learning', 'study', 'results'}
        # filtered_pairs = {pair: freq for pair, freq in cooccurrence_dict.items() if not common_filter.intersection(pair)}
        # top_pairs = sorted(filtered_pairs.items(), key=lambda item: item[1], reverse=True)[:top_N_pairs]
        top_pairs = sorted(cooccurrence_dict.items(), key=lambda item: item[1], reverse=True)[:top_N_pairs]


        if top_pairs:
            logging.info("Generating co-occurrence plot...")
            # Prepare for plotting
            pairs_labels = [f"{pair[0]}-{pair[1]}" for pair, freq in top_pairs]
            pairs_values = [freq for pair, freq in top_pairs]

            plt.figure(figsize=(12, 10)) # Adjusted size
            plt.barh(pairs_labels[::-1], pairs_values[::-1], color='salmon') # Invert order for plot
            plt.xlabel('Co-occurrence Frequency')
            plt.ylabel('Keyword Pair')
            plt.title(f'Top {top_N_pairs} Co-occurring Keyword Pairs in Results', fontsize=16, fontweight='bold')
            plt.tight_layout()
            plt.savefig(FIGURES_DIR / 'coocurrencia_terminos.png', dpi=300, bbox_inches='tight')
            logging.info(f"Saved term co-occurrence plot to {FIGURES_DIR / 'coocurrencia_terminos.png'}")
            plt.show()
        else:
             logging.warning("No top co-occurring pairs found after filtering.")
    else:
        logging.warning("No co-occurring terms found.")
else:
    logging.warning("Skipping co-occurrence analysis due to missing 'General Results' or word frequencies.")

# %% [markdown]
# ### 17. Sample Size Analysis

# %%
# --- Sample Size Analysis ---
if 'Sample Size' in df.columns:
    logging.info("Performing sample size analysis...")
    sample_sizes = []
    # Improved regex to handle ranges and single numbers more robustly
    pattern = re.compile(r'\d+') # Find all sequences of digits

    for size_str in df['Sample Size'].dropna().astype(str):
        numbers = pattern.findall(size_str)
        if numbers:
            # If multiple numbers found (like a range '30-40'), take the average or max? Let's take max for now.
            numeric_values = [int(n) for n in numbers]
            sample_sizes.append(max(numeric_values))
            # Alternative: average for ranges
            # if len(numeric_values) == 2 and '-' in size_str:
            #     sample_sizes.append(sum(numeric_values) / 2)
            # elif numeric_values:
            #     sample_sizes.append(numeric_values[0]) # Take first number if not clearly a range

    if sample_sizes:
        # Calculate statistics
        sample_stats = {
            'Mean': np.mean(sample_sizes),
            'Median': np.median(sample_sizes),
            'Std Dev': np.std(sample_sizes),
            'Min': np.min(sample_sizes),
            'Max': np.max(sample_sizes),
            'Count': len(sample_sizes) # Number of studies with sample size info
        }

        print("\nSample Size Statistics:")
        for stat, value in sample_stats.items():
            print(f"- {stat}: {value:.1f}")

        # Visualize distribution
        logging.info("Generating sample size distribution plot...")
        plt.figure(figsize=(12, 7))
        # Use more bins if range is large, limit x-axis if outliers distort
        max_val_display = np.percentile(sample_sizes, 98) # Exclude top 2% for better visualization scale
        plt.hist(sample_sizes, bins=30, color='skyblue', edgecolor='navy', range=(0, max_val_display))
        plt.axvline(sample_stats['Mean'], color='red', linestyle='--', label=f'Mean: {sample_stats["Mean"]:.1f}')
        plt.axvline(sample_stats['Median'], color='green', linestyle='--', label=f'Median: {sample_stats["Median"]:.1f}')
        plt.xlabel('Sample Size (Max reported or average if range)')
        plt.ylabel('Frequency (Number of Studies)')
        plt.title('Distribution of Sample Sizes', fontsize=16, fontweight='bold')
        plt.legend()
        plt.grid(True, axis='y', linestyle='--', alpha=0.6)
        plt.tight_layout()
        plt.savefig(FIGURES_DIR / 'distribucion_tamano_muestra.png', dpi=300, bbox_inches='tight')
        logging.info(f"Saved sample size distribution plot to {FIGURES_DIR / 'distribucion_tamano_muestra.png'}")
        plt.show()
    else:
        logging.warning("Could not extract numeric sample size data from the 'Sample Size' column.")
else:
    logging.warning("Column 'Sample Size' not found. Skipping sample size analysis.")

# %% [markdown]
# ### 18. Age/Grade Analysis

# %%
# --- Age/Grade Analysis ---
if 'Age/Grade' in df.columns:
    logging.info("Performing Age/Grade analysis...")
    # Define more specific patterns or keywords
    # This part is highly dependent on the specific format in your CSV
    age_grade_categories = defaultdict(int)

    for age_str in df['Age/Grade'].dropna().astype(str):
        age_lower = age_str.lower().strip()
        matched = False
        # Try to match specific grades first
        grade_match = re.search(r'(\d+)(?:st|nd|rd|th)?\s+grade', age_lower)
        if grade_match:
             # You could map grades to age groups or keep them separate
             age_grade_categories[f'Grade {grade_match.group(1)}'] += 1
             matched = True
        # Try to match age ranges
        elif '-' in age_lower or '' in age_lower: # Handle different dashes
             age_grade_categories['Age Range Specified'] += 1
             matched = True
        # Try to match specific ages
        elif re.search(r'\d+\s+years', age_lower) or re.search(r'age\s+\d+', age_lower):
             age_grade_categories['Specific Age(s) Specified'] += 1
             matched = True
        elif age_lower in ['primary', 'elementary']:
             age_grade_categories['General (Primary/Elementary)'] += 1
             matched = True

        if not matched and age_lower: # Catch non-empty strings that didn't match
            age_grade_categories['Other Description'] += 1
        elif not age_lower: # Count empty strings if needed
             age_grade_categories['Not Specified/Empty'] += 1


    # Add count for originally NaN values
    age_grade_categories['Not Specified/Empty'] += df['Age/Grade'].isnull().sum()

    # Filter out categories with zero counts
    age_patterns_plot = {k:v for k,v in age_grade_categories.items() if v > 0}


    if age_patterns_plot:
        logging.info("Generating age/grade patterns plot...")
        # Visualize patterns
        plt.figure(figsize=(12, max(6, len(age_patterns_plot) * 0.5))) # Dynamic height
        age_df = pd.DataFrame.from_dict(age_patterns_plot, orient='index', columns=['Frequency'])
        age_df = age_df.sort_values('Frequency', ascending=True)
        bars = plt.barh(age_df.index, age_df['Frequency'], color='lightcoral') # Changed color
        plt.xlabel('Number of Studies')
        plt.ylabel('Age/Grade Description Pattern')
        plt.title('Patterns in Age/Grade Descriptions', fontsize=16, fontweight='bold')

        # Add labels
        for bar in bars:
            width = bar.get_width()
            plt.text(width + 0.1, bar.get_y() + bar.get_height()/2., f'{int(width)}',
                     ha='left', va='center', fontsize=10)

        plt.tight_layout()
        plt.savefig(FIGURES_DIR / 'patrones_edad_grado.png', dpi=300, bbox_inches='tight')
        logging.info(f"Saved age/grade patterns plot to {FIGURES_DIR / 'patrones_edad_grado.png'}")
        plt.show()
    else:
         logging.warning("No age/grade patterns found or column missing.")
else:
    logging.warning("Column 'Age/Grade' not found. Skipping age/grade analysis.")


# %% [markdown]
# ### 19. Complex Relationship Matrix (Heatmap)

# %%
# --- Relationship Matrix ---
# Requires Intervention classification, Methodology columns, Sentiment column
matrix_cols_exist = ('Intervention' in df.columns and
                     'Quantitative Methods' in df.columns and
                     'Qualitative Methods' in df.columns and
                     'Mixed Methods' in df.columns and
                     'Sentiment' in df.columns and # Assumes Sentiment was calculated in Step 13
                     'intervention_keywords' in locals()) # Need keywords for classification

if matrix_cols_exist:
    logging.info("Generating complex relationship matrix heatmap...")
    # Define primary intervention types for matrix rows
    matrix_intervention_types = list(intervention_keywords.keys()) # Use keys from defined dict

    # Define matrix columns
    matrix_columns = ['Quantitative', 'Qualitative', 'Mixed', 'Positive', 'Negative', 'Neutral']

    # Initialize DataFrame with zeros
    relationship_matrix = pd.DataFrame(0, index=matrix_intervention_types, columns=matrix_columns)

    # Populate the matrix
    for _, row in df.iterrows():
        intervention_type = None
        # Classify intervention
        if pd.notna(row['Intervention']):
             intervention_lower = row['Intervention'].lower()
             for type_name, keywords in intervention_keywords.items():
                 if any(term in intervention_lower for term in keywords):
                     intervention_type = type_name
                     break

        if intervention_type and intervention_type in relationship_matrix.index: # Check if it's a type we are tracking
            # Increment methodology counts
            if pd.notna(row['Quantitative Methods']): relationship_matrix.loc[intervention_type, 'Quantitative'] += 1
            if pd.notna(row['Qualitative Methods']): relationship_matrix.loc[intervention_type, 'Qualitative'] += 1
            if pd.notna(row['Mixed Methods']): relationship_matrix.loc[intervention_type, 'Mixed'] += 1
            # Increment sentiment counts
            sentiment = row.get('Sentiment', 'Neutral') # Get sentiment calculated earlier
            if sentiment in relationship_matrix.columns: # Check if sentiment category is a column
                 relationship_matrix.loc[intervention_type, sentiment] += 1

    # Convert to integer type
    relationship_matrix = relationship_matrix.astype(int)

    # Filter out rows/columns with all zeros if desired for visual clarity
    relationship_matrix = relationship_matrix.loc[(relationship_matrix.sum(axis=1) > 0), (relationship_matrix.sum(axis=0) > 0)]

    if not relationship_matrix.empty:
        # Visualize heatmap
        plt.figure(figsize=(12, 8)) # Adjusted size
        sns.heatmap(relationship_matrix, annot=True, fmt='d', cmap='YlGnBu', # Changed colormap
                    linewidths=.5, linecolor='grey',
                    cbar_kws={'label': 'Number of Studies'}, annot_kws={"size": 10}) # Adjust annotation size
        plt.title('Relationship Matrix: Interventions vs. Methodology & Results',
                  fontsize=15, fontweight='bold') # Adjusted size
        plt.xlabel('Methodology / Result Sentiment')
        plt.ylabel('Intervention Type')
        plt.xticks(rotation=45, ha='right')
        plt.yticks(rotation=0)
        plt.tight_layout()
        plt.savefig(FIGURES_DIR / 'matriz_relaciones_complejas.png', dpi=300, bbox_inches='tight')
        logging.info(f"Saved relationship matrix heatmap to {FIGURES_DIR / 'matriz_relaciones_complejas.png'}")
        plt.show()
    else:
         logging.warning("Relationship matrix is empty after processing. Skipping heatmap.")

else:
    logging.warning("Required columns for relationship matrix are missing. Skipping heatmap.")

# %% [markdown]
# ### 20. Final Summary and Index Generation

# %%
logging.info("Generating final summary and index file...")

# --- Final Conclusions Radar Plot (Optional - Needs careful interpretation) ---
# Check if necessary calculated variables exist
radar_data_exists = ('methods_analysis' in locals() and methods_analysis and
                     'intervention_counts' in locals() and intervention_counts and
                     'sentiment_distribution' in locals() and not sentiment_distribution.empty and
                     'theme_counts' in locals() and theme_counts and
                     'sample_sizes' in locals() and sample_sizes and
                     'country_counts' in locals() and not country_counts.empty)

if radar_data_exists:
    logging.info("Generating summary radar plot...")
    # Create a summary visualization (e.g., Radar Chart summarizing key aspects)
    categories = ['Geographic\nDiversity', 'Mixed\nMethods Use', 'AR/VR\nPrevalence',
                  'Positive\nResults Rate', 'Thematic\nCoverage', 'Sample Size\nReporting']
    # Normalize values (simple normalization, adjust scales as needed)
    total_studies = len(df)
    max_countries_expected = 30 # Arbitrary max for normalization
    max_keywords_expected = 1000 # Arbitrary max for normalization

    values = [
        min(country_counts.count() / max_countries_expected * 100, 100), # Diversity based on number of unique countries
        methods_analysis.get('Mixed', 0) / total_studies * 100, # % Mixed Methods
        intervention_counts.get('AR/VR', 0) / total_studies * 100, # % AR/VR
        sentiment_distribution.get('Positive', 0) / total_studies * 100, # % Positive Sentiment
        min(sum(theme_counts.values()) / max_keywords_expected * 100, 100), # Thematic coverage (normalized)
        len(sample_sizes) / total_studies * 100 # % Studies reporting usable sample size
    ]

    # Complete the loop for radar chart
    angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
    values += values[:1]
    angles += angles[:1]

    fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))
    ax.plot(angles, values, 'o-', linewidth=2, color='dodgerblue')
    ax.fill(angles, values, alpha=0.25, color='dodgerblue')
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories, fontsize=10) # Adjust font size
    ax.set_yticks(np.arange(0, 101, 20)) # Y-ticks from 0 to 100
    ax.set_ylim(0, 100)
    ax.set_title('Systematic Review Profile (Normalized Indicators)', size=16, weight='bold', pad=25)

    # Add percentage labels
    for angle, value in zip(angles[:-1], values[:-1]):
        ax.text(angle, value + 5, f'{value:.0f}%', ha='center', va='center', fontweight='bold', fontsize=9)

    plt.tight_layout()
    plt.savefig(FIGURES_DIR / 'perfil_revision_sistematica.png', dpi=300, bbox_inches='tight')
    logging.info(f"Saved summary radar plot to {FIGURES_DIR / 'perfil_revision_sistematica.png'}")
    plt.show()
else:
    logging.warning("Could not generate summary radar plot due to missing prerequisite data.")


# --- Generate Index File ---
logging.info("Creating index file for generated figures...")
figure_files_list = sorted(FIGURES_DIR.glob('*.*')) # Get all files (png and html)
index_content = "# Index of Generated Figures\n\n"
index_content += "This directory contains the figures generated by the analysis script.\n\n"

if figure_files_list:
    for i, f_path in enumerate(figure_files_list):
        # Add a brief description based on filename (can be improved)
        description = f_path.stem.replace('_', ' ').capitalize()
        index_content += f"{i+1}. **{f_path.name}**: {description}\n"
else:
    index_content += "*No figure files were found in the directory.*\n"

index_file_path = FIGURES_DIR / 'AAA_FIGURE_INDEX.md' # Start with AAA for sorting
try:
    with open(index_file_path, 'w', encoding='utf-8') as f:
        f.write(index_content)
    logging.info(f"Figure index file created at: {index_file_path}")
except Exception as e:
    logging.error(f"Failed to write figure index file: {e}")


# --- Final Console Summary ---
print("\n" + "=" * 60)
print("      ANALYSIS AND FIGURE GENERATION COMPLETE")
print("=" * 60)
print(f"Date and Time: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"Output Directory for Figures: {FIGURES_DIR.resolve()}")
print(f"\nTotal figures generated: {len(figure_files_list)}")
if figure_files_list:
    print("Generated files include:")
    for f_path in figure_files_list:
        print(f"  - {f_path.name}")
print(f"\nAn index file 'AAA_FIGURE_INDEX.md' has been created in the output directory.")
print("=" * 60)
```

```python
# %%
# --- Imports ---
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches # Needed for legend
import numpy as np # Needed for calculating positions
import logging
from pathlib import Path

# --- Setup (Assume these are defined earlier in your script) ---
# logging.basicConfig(level=logging.INFO)
# df = pd.read_csv(...) # Your DataFrame
# FIGURES_DIR = Path("./figures")
# FIGURES_DIR.mkdir(exist_ok=True)
# Example intervention keywords (replace or ensure this is defined earlier)
# intervention_keywords = {
#     'AR/VR': ['ar', 'vr', 'augmented reality', 'virtual reality'],
#     'Games/Gamification': ['game', 'gamification', 'serious game'],
#     'Educational Software': ['software', 'platform', 'tool'],
#     'Adaptive System': ['adaptive', 'personalized', 'intelligent tutoring'],
#     'Robotics': ['robot', 'robotics']
# }
# Make sure df and FIGURES_DIR are loaded/defined before this cell runs

# %%
# --- Network Analysis ---
required_net_cols = ['FirstCountry', 'Quantitative Methods', 'Qualitative Methods', 'Mixed Methods', 'Intervention']

# Placeholder for missing columns check if needed
# if not all(col in df.columns for col in required_net_cols):
#     missing_cols = [col for col in required_net_cols if col not in df.columns]
#     logging.error(f"Required columns missing for network analysis: {missing_cols}. Stopping.")
#     # Or handle the error appropriately, maybe exit or skip the section
#     # exit() # Uncomment to stop execution if columns are missing

if all(col in df.columns for col in required_net_cols):
    logging.info("Generating relationship network graph...")
    G = nx.Graph() # Undirected graph for co-occurrence

    # Define node types and colors
    # Using slightly different colors for better visual distinction if needed
    node_colors_map = {'method': 'mediumseagreen', 'intervention': 'lightcoral', 'country': 'lightskyblue'}

    # --- Node Addition (Same as your code) ---
    # Add Country nodes
    countries = df['FirstCountry'].dropna().unique()
    for country in countries:
        G.add_node(country, node_type='country')

    # Add Intervention nodes
    # Check if 'intervention_keywords' exists in the global or local scope
    intervention_keywords_available = 'intervention_keywords' in locals() or 'intervention_keywords' in globals()
    if intervention_keywords_available:
        # Ensure 'intervention_keywords' is accessible
        if 'intervention_keywords' not in locals() and 'intervention_keywords' in globals():
             intervention_keywords = globals()['intervention_keywords']
        intervention_types_unique = list(intervention_keywords.keys())
        for intervention in intervention_types_unique:
             G.add_node(intervention, node_type='intervention')
        G.add_node('Other Intervention', node_type='intervention')
    else:
        logging.warning("Intervention keywords not defined; using hardcoded list for network.")
        intervention_types_unique = ['AR/VR', 'Games/Gamification', 'Educational Software', 'Adaptive System', 'Robotics', 'Other Intervention']
        for intervention in intervention_types_unique:
            G.add_node(intervention, node_type='intervention')

    # Add Method nodes
    method_types = ['Quantitative', 'Qualitative', 'Mixed']
    for method in method_types:
        G.add_node(method, node_type='method')

    # --- Edge Addition (Corrected Method Checks) ---
    for _, row in df.iterrows():
        country = row['FirstCountry']
        # Skip if country is NaN or the country node wasn't added (e.g., empty string initially)
        if pd.isna(country) or not G.has_node(country):
             continue

        # Determine intervention type for the row
        row_intervention_node = None
        if intervention_keywords_available and pd.notna(row['Intervention']): # Check intervention is not NaN
            intervention_lower = str(row['Intervention']).lower() # Convert to string just in case
            for type_name, keywords in intervention_keywords.items():
                 if any(str(term) in intervention_lower for term in keywords): # Ensure terms are strings
                    if G.has_node(type_name):
                        row_intervention_node = type_name
                        break
            if row_intervention_node is None and G.has_node('Other Intervention'):
                row_intervention_node = 'Other Intervention'

        # Link Country to Intervention Type
        if row_intervention_node:
            # Ensure the country node exists before adding edge
            if G.has_node(country):
                G.add_edge(country, row_intervention_node)
            # else: # Optional logging if country node was missing
            #    logging.debug(f"Country node '{country}' not found for linking intervention '{row_intervention_node}'.")


        # Determine methods for the row using pd.notna()
        row_method_nodes = []
        # *** CORRECTION HERE ***
        # Check if the value in the column is not null/NaN
        if pd.notna(row['Quantitative Methods']) and G.has_node('Quantitative'):
             row_method_nodes.append('Quantitative')
        if pd.notna(row['Qualitative Methods']) and G.has_node('Qualitative'):
             row_method_nodes.append('Qualitative')
        if pd.notna(row['Mixed Methods']) and G.has_node('Mixed'):
             row_method_nodes.append('Mixed')
        # *** END CORRECTION ***


        # Link Country to Methodology
        if G.has_node(country): # Ensure country node exists
            for method_node in row_method_nodes:
                 G.add_edge(country, method_node)
        # else: # Optional logging
        #     logging.debug(f"Country node '{country}' not found for linking methods.")


        # Link Intervention to Methodology (Optional part)
        if row_intervention_node: # Ensure intervention node exists
             for method_node in row_method_nodes: # Ensure method node exists (implicit from above)
                 G.add_edge(row_intervention_node, method_node)

    # Remove isolated nodes
    isolated = list(nx.isolates(G))
    if isolated:
        G.remove_nodes_from(isolated)
        logging.info(f"Removed {len(isolated)} isolated nodes.")

    # --- Visualization with Concentric Layout and Legend ---
    if G.number_of_nodes() > 0 and G.number_of_edges() > 0:
        plt.figure(figsize=(24, 24)) # Make figure larger for this layout

        # 1. Separate nodes by type
        method_nodes = [n for n, d in G.nodes(data=True) if d.get('node_type') == 'method']
        intervention_nodes = [n for n, d in G.nodes(data=True) if d.get('node_type') == 'intervention']
        country_nodes = [n for n, d in G.nodes(data=True) if d.get('node_type') == 'country']

        # 2. Calculate positions for concentric ellipses (using circles here for simplicity)
        pos = {}
        center_radius = 0.2 # Radius for method nodes (close to center)
        intervention_radius = 0.35 # Radius for intervention nodes
        country_radius = 0.55   # Radius for country nodes

        # Position Method nodes (innermost)
        num_methods = len(method_nodes)
        if num_methods > 0:
            angles_m = np.linspace(0, 2 * np.pi, num_methods, endpoint=False)
            for i, node in enumerate(method_nodes):
                 # Add slight variation if only one node to avoid exact center
                r = center_radius if num_methods > 1 else 0.01
                pos[node] = (r * np.cos(angles_m[i]), r * np.sin(angles_m[i]))

        # Position Intervention nodes (middle)
        num_interventions = len(intervention_nodes)
        if num_interventions > 0:
            angles_i = np.linspace(0, 2 * np.pi, num_interventions, endpoint=False)
             # Start angle slightly offset for better separation from methods if needed
            start_angle_i = np.pi / num_interventions if num_interventions > 0 else 0
            for i, node in enumerate(intervention_nodes):
                pos[node] = (intervention_radius * np.cos(angles_i[i] + start_angle_i),
                             intervention_radius * np.sin(angles_i[i] + start_angle_i))

        # Position Country nodes (outermost)
        num_countries = len(country_nodes)
        if num_countries > 0:
            angles_c = np.linspace(0, 2 * np.pi, num_countries, endpoint=False)
             # Start angle slightly offset
            start_angle_c = np.pi / num_countries if num_countries > 0 else 0
            for i, node in enumerate(country_nodes):
                pos[node] = (country_radius * np.cos(angles_c[i] + start_angle_c),
                             country_radius * np.sin(angles_c[i] + start_angle_c))

        # Check if all nodes got a position (debugging)
        if len(pos) != G.number_of_nodes():
             logging.warning(f"Position calculated for {len(pos)} nodes, but graph has {G.number_of_nodes()} nodes.")
             # Fallback to spring layout if positioning failed for some nodes
             missing_nodes = [n for n in G.nodes() if n not in pos]
             logging.warning(f"Nodes without calculated position: {missing_nodes}")
             logging.info("Falling back to spring layout.")
             pos = nx.spring_layout(G, k=0.6, iterations=50, seed=78)


        # 3. Get node colors and sizes
        node_colors = [node_colors_map.get(G.nodes[node].get('node_type', 'unknown'), 'grey') for node in G.nodes()]
        degrees = dict(G.degree())
        # Adjust size formula - may need smaller nodes for circular layout
        # Start with a moderate size and adjust based on visual output
        node_sizes = [degrees.get(node, 1) * 650 + 600 for node in G.nodes()]

        # 4. Draw the network elements
        nx.draw_networkx_edges(G, pos, alpha=0.2, edge_color='grey', width=3.0) # Thinner edges often better here
        nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=node_sizes, alpha=0.9)
        nx.draw_networkx_labels(G, pos, font_size=18, font_weight='normal') # Smaller font often needed

        # 5. Add the Legend
        legend_handles = [mpatches.Patch(color=color, label=label.replace('_', ' ').capitalize())
                          for label, color in node_colors_map.items()]
        plt.legend(handles=legend_handles, title="Node Types", loc='upper right', fontsize=20, title_fontsize='large')


        #plt.title('Co-occurrence Network (Concentric Layout)', fontsize=24, fontweight='bold')
        plt.axis('off')
        plt.tight_layout() # Adjust layout to prevent title/legend overlap if possible

        # Adjust plot limits to ensure the legend fits if it's outside axes
        plt.margins(0.1) # Add some margin around the plot

        # Save the figure
        # Ensure FIGURES_DIR is defined and is a Path object or string
        if 'FIGURES_DIR' in locals() or 'FIGURES_DIR' in globals():
            if 'FIGURES_DIR' not in locals() and 'FIGURES_DIR' in globals():
                 FIGURES_DIR = globals()['FIGURES_DIR']
            save_path = Path(FIGURES_DIR) / 'red_relaciones_concentrica.png'
            plt.savefig(save_path, dpi=300, bbox_inches='tight')
            logging.info(f"Saved concentric relationship network plot to {save_path}")
        else:
            logging.warning("FIGURES_DIR not defined. Cannot save plot.")

        plt.show()

    elif G.number_of_nodes() == 0:
         logging.warning("Network graph has no nodes after processing. Skipping visualization.")
    else: # Has nodes but no edges
         logging.warning("Network graph has nodes but no edges after processing. Skipping visualization.")


else:
    missing_cols = [col for col in required_net_cols if col not in df.columns]
    logging.warning(f"Required columns missing for network analysis: {missing_cols}. Skipping network graph.")
```
